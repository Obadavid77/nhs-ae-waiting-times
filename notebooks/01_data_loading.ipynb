### Objective
Load all 36 monthly CSV files from NHS England covering April 2023 to 
February 2026. Combine them into a single dataset and perform an initial 
inspection to understand the structure, size, and quality of the data 
before any cleaning takes place.

### Data Source
NHS England — A&E Attendances and Emergency Admissions  
Licence: Open Government Licence v3.0  
URL: https://www.england.nhs.uk/statistics/statistical-work-areas/ae-waiting-times-and-activity/

```python
import pandas as pd
import glob
import os

# Path to your data folder
folder_path = '../data/raw/'

# Load all CSV files
all_files = glob.glob(os.path.join(os.path.dirname(os.getcwd()), 'data', 'raw', '*.csv'))

print(f'Number of CSV files found: {len(all_files)}')
print('\nFiles found:')
for f in all_files:
    print(os.path.basename(f))
```

```python
import os

# Check where Jupyter thinks we are right now
print(os.getcwd())
```

```python
import pandas as pd
import glob
import os

# Load all 36 files and combine into one dataframe
dfs = []

for file in all_files:
    df = pd.read_csv(file)
    # Add a column so we know which file each row came from
    df['source_file'] = os.path.basename(file)
    dfs.append(df)

# Combine everything into one table
combined = pd.concat(dfs, ignore_index=True)

print(f'Total rows: {len(combined)}')
print(f'Total columns: {len(combined.columns)}')
print(f'\nColumn names:')
for col in combined.columns:
    print(col)
```

```python
# Look at the first 5 rows to understand the data
print("First 5 rows:")
combined.head()
```

```python
# Check for missing values in our key columns
key_cols = [
    'Period',
    'Org name',
    'A&E attendances Type 1',
    'Attendances over 4hrs Type 1'
]

print("Missing values in key columns:")
print(combined[key_cols].isnull().sum())

print(f"\nDate range in Period column:")
print(combined['Period'].unique()[:10])
```

```python
# Step 1 - Remove total/summary rows (not real Trust data)
print(f"Rows before removing totals: {len(combined)}")

combined = combined[~combined['Period'].astype(str).str.contains('TOTAL|Total', na=False)]

print(f"Rows after removing totals: {len(combined)}")

# Step 2 - Keep only real MSitAE period rows
combined = combined[combined['Period'].astype(str).str.startswith('MSitAE')]

print(f"Rows after keeping only valid periods: {len(combined)}")

# Step 3 - Convert Period to a proper date
# MSitAE-APRIL-2025 -> April 2025
combined['Period_clean'] = combined['Period'].str.replace('MSitAE-', '', regex=False)
combined['Period_date'] = pd.to_datetime(combined['Period_clean'], format='%B-%Y')

# Step 4 - Drop junk columns
cols_to_drop = [col for col in combined.columns if 'Unnamed' in str(col) or col == 'a']
combined = combined.drop(columns=cols_to_drop)

# Step 5 - Extract month and year
combined['month'] = combined['Period_date'].dt.month
combined['year'] = combined['Period_date'].dt.year

print(f"\nDate range: {combined['Period_date'].min()} to {combined['Period_date'].max()}")
print(f"\nFinal shape: {combined.shape}")
print(f"\nSample of cleaned periods:")
print(combined[['Period', 'Period_date', 'month', 'year']].drop_duplicates().sort_values('Period_date').head(10))
```

```python
# Step 6 - Convert key columns to numeric (some may have dashes or blanks)
combined['A&E attendances Type 1'] = pd.to_numeric(
    combined['A&E attendances Type 1'], errors='coerce')

combined['Attendances over 4hrs Type 1'] = pd.to_numeric(
    combined['Attendances over 4hrs Type 1'], errors='coerce')

# Step 7 - Create our target variable: breach rate
combined['breach_rate'] = (
    combined['Attendances over 4hrs Type 1'] / 
    combined['A&E attendances Type 1']
) * 100

# Step 8 - Remove rows where attendances are 0 or missing
# (can't calculate breach rate without attendances)
before = len(combined)
combined = combined[combined['A&E attendances Type 1'] > 0]
combined = combined.dropna(subset=['breach_rate'])
after = len(combined)

print(f"Rows removed due to zero/missing attendances: {before - after}")
print(f"Final usable rows: {after}")

# Step 9 - Sanity check the breach rate
print(f"\nBreach rate statistics:")
print(combined['breach_rate'].describe().round(2))

print(f"\nAny breach rates over 100% (impossible):")
print(len(combined[combined['breach_rate'] > 100]))

print(f"\nSample of data with breach rate:")
print(combined[['Period_date', 'Org name', 
                 'A&E attendances Type 1',
                 'Attendances over 4hrs Type 1',
                 'breach_rate']].head(10).round(2))
```

```python
# Save the cleaned dataset
output_path = '../data/processed/'
os.makedirs(output_path, exist_ok=True)

combined.to_csv(output_path + 'nhs_ae_clean.csv', index=False)

print(f"Clean dataset saved successfully.")
print(f"Location: data/processed/nhs_ae_clean.csv")
print(f"Shape: {combined.shape}")
print(f"Columns saved:")
for col in combined.columns:
    print(f"  - {col}")
```
